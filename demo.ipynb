```python
#import cv2
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from IttiKoch import IttiKoch
```

```python
img1path = 'sample_images/aircrafts.png'
img2path = 'sample_images/balloons.png'
img3path = 'sample_images/faces.png'

img1 = np.array( Image.open( img1path ) )
img2 = np.array( Image.open( img2path ) )
img3 = np.array( Image.open( img3path ) )

itti = IttiKoch()

scan1 = itti.getScanPath( img1path, saliency_type = '01' )
sm1 = itti.SM_fullSize
scan2 = itti.getScanPath( img2path, saliency_type = '01' )
sm2 = itti.SM_fullSize
scan3 = itti.getScanPath( img3path, saliency_type = '01' )
sm3 = itti.SM_fullSize
```

```python
# og images

plt.subplot(3,3,1)
plt.imshow( img1 )
plt.axis('off')

plt.subplot(3,3,2)
plt.imshow( img2 )
plt.axis('off')

plt.subplot(3,3,3)
plt.imshow( img3 )
plt.axis('off')

# saleincy maps

plt.subplot(3,3,4)
plt.imshow( sm1 )
plt.axis('off')

plt.subplot(3,3,5)
plt.imshow( sm2 )
plt.axis('off')

plt.subplot(3,3,6)
plt.imshow( sm3 )
plt.axis('off')

# scanpaths

plt.subplot(3,3,7)
plt.imshow( img1 )
plt.plot(scan1[:,1], scan1[:,0], '-yo')
plt.axis('off')

plt.subplot(3,3,8)
plt.imshow( img2 )
plt.plot(scan2[:,1], scan2[:,0], '-yo')
plt.axis('off')

plt.subplot(3,3,9)
plt.imshow( img3 )
plt.plot(scan3[:,1], scan3[:,0], '-yo')
plt.axis('off')
```
